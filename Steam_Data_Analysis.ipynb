```python
#sorgu 1: Rating'i very positive olan ve final price değeri 10.00'ın üzerinde olan oyunların id'lerini ve isimlerini bastır

filt = (df["rating"]=="Very Positive" ) & (df['price_final'] > 10.00)
df.loc[filt, ["app_id","title"]]
```

```python
# sorgu 2: Linux'ta çalışmaya uyumlu ve discount değeri 0'dan farklı olan oyunların isimlkerini, ve final fiyatlarını listele

filt = ((df["linux"]==True) & (df["discount"] != 0))

df.loc[filt, ["title", "price_final"]]
```

```python
# sorgu 3: 2021 yılında çıkmış, steam deck'e uyumlu ve ;
### ÖNEMLİ
```

```python
filt = ((((df['date_release'].apply(lambda x: int(x.split('-')[0])) == 2021  )) & (df["steam_deck"]==True)) )
df.loc[filt, ['title', 'date_release']]
```

```python
df['date_release'].apply(lambda x: x.split('-')[0]) == 2021
```

```python
# sorgu 4: # ya positive ratio'su 60'dan
# büyük olan ya da price değeri 4.99'dan fazla olan oyunları listele

filt = (df["positive_ratio"] > 60) | (df["price_final"] > 4.99)
df.loc[filt, ["title","date_release", "price_final" ,"positive_ratio"]]
```
